## American discretisation 

Crank Nicolson gives:

$$
a_jB_{j-1}^i + b_j B_j^i + c_jB^i_{j+1}=d_j
$$

where:

$$
a_j = \frac{\sigma^2 (j\Delta r)^{2\beta}}{4(\Delta r)^2} - \frac{\kappa(\theta e^{\mu i\Delta t}-j\Delta r)}{4\Delta r}
$$
$$
b_j = -\frac{1}{\Delta t} - \frac{\sigma^2(j\Delta r)^{2\beta}}{2(\Delta r)^2} - \frac{1}{2}j\Delta r
$$
$$
c_j = \frac{\sigma^2 (j\Delta r)^{2\beta}}{4(\Delta r)^2} + \frac{\kappa(\theta e^{\mu i\Delta t}-j\Delta r)}{4\Delta r}
$$
$$
d_j = -V^{i+1}_{j-1}\left[\frac{\sigma^2 (j\Delta r)^{2\beta}}{4(\Delta r)^2} - \frac{\kappa(\theta e^{\mu i\Delta t}-j\Delta r)}{4\Delta r}\right]
$$
$$
-V^{i+1}_j\left[\frac{1}{\Delta t} - \frac{\sigma^2(j\Delta r)^{2\beta}}{2(\Delta r)^2} - \frac{1}{2}j\Delta r\right]
$$
$$
-V^{i+1}_{j-1}\left[\frac{\sigma^2 (j\Delta r)^{2\beta}}{4(\Delta r)^2} + \frac{\kappa(\theta e^{\mu i\Delta t}-j\Delta r)}{4\Delta r}\right]
$$

Boundary LHS

$$
\begin{align}
\nonumber
a_0 &= -\frac{1}{\Delta t} - \frac{3\kappa\theta e^{\mu i\Delta t}}{4\Delta r} \\
\nonumber
b_0 &= \frac{\kappa\theta e^{\mu i\Delta t}}{\Delta r}\\
\nonumber
c_0 &= -\frac{\kappa\theta e^{\mu i\Delta t}}{4\Delta r}\\
\nonumber
d_0 &= -V_0^{i+1}\left[\frac{1}{\Delta t} - \frac{3\kappa\theta e^{\mu i\Delta t}}{4\Delta r}\right]\\
\nonumber
&-V_1^{i+1}\left[\frac{\kappa\theta e^{\mu i\Delta t}}{\Delta r}\right]\\
\nonumber
&-V_2^{i+1}\left[ -\frac{\kappa\theta e^{\mu i\Delta t}}{4\Delta r}\right]
\end{align}
$$

$$
\begin{align}
\nonumber
a_1 &= \frac{\sigma^2(\Delta r)^{2\beta}}{4(\Delta r)^2} - \frac{\kappa(\theta e^{\mu i\Delta t}-\Delta r)}{4\Delta r}\\
\nonumber
b_1 &= -\frac{1}{\Delta t} - \frac{\sigma^2(\Delta r)^{2\beta}}{2(\Delta r)^2} - \frac{1}{2}\Delta r\\
\nonumber
c_1 &= \frac{\sigma^2(\Delta r)^{2\beta}}{4(\Delta r)^2} + \frac{\kappa(\theta e^{\mu i\Delta t}-\Delta r)}{4\Delta r}\\
\nonumber
d_1 &= -V_0^{i+1}\left[\frac{\sigma^2(\Delta r)^{2\beta}}{4(\Delta r)^2} - \frac{\kappa(\theta e^{\mu i\Delta t}-\Delta r)}{4\Delta r}\right]\\
\nonumber
& -V_1^{i+1}\left[\frac{1}{\Delta t} - \frac{\sigma^2(\Delta r)^{2\beta}}{2(\Delta r)^2} - \frac{1}{2}\Delta r\right]\\
\nonumber
& -V_2^{i+1}\left[\frac{\sigma^2(\Delta r)^{2\beta}}{4(\Delta r)^2} - \frac{\kappa(\theta e^{\mu i\Delta t}-\Delta r)}{4\Delta r}\right]

\end{align}
$$

and finally we implement:

$$
\begin{align}
\tilde{a}_0 &= 0\\
\tilde{b}_0 &= a_0 - \frac{c_0}{c_1}a_1\\
\tilde{c}_0 &= b_0 - \frac{c_0}{c_1}b_1\\
\tilde{d}_0 &= d_0 - \frac{c_0}{c_1}d_1

\end{align}
$$

Boundary RHS:

$$
\begin{align}
a_{j_{max}}&=0\\

b_{j_{max}} &= 1

c_{j_{max}} &= 0\\

d_{j_{max}} &= X - B^i_{j_{max}}

\end{align}
$$
